**<font color = blue > 1. Explain supervised machine learning, the notion of labelled data, and train and test datasets. </font>**

**<font color = red > Supervised Machine learning is a type of machine learning where the model is trained on a labeled dataset. Each example in the dataset has an input data and a corresponding output label and the objective of the model is to determine the relationship between the input data and the output labels so that it can make predictions on the new and unseen data. Regression and classification are the two tasks in supervised machine learning.The objective of the model is to predict a continuous value or quantity in regression and a categorical label and class in classification.</font>**

**<font color = red > Labelled data is a dataset where each instance has a corresponding output label.Labelled data provides the necessary information for supervised machine learning algorithms to understand the relationship between input features and their corresponding outputs. Without labelled data, Algorithms fail to learn meaningful patterns  By using labelled data, models can generalize their learnings to new, unseen examples, improving their predictiions.</font>**

**<font color = red >In supervised machine learning a dataset is usually divided into two subsets that are known as training dataset and test dataset. Training dataset is used to train the machine learning model.The data consists of input and output values.The model makes use of this training dataset to learn the relationship between the input values and the output labels. Once the model is trained using the training dataset its effienciey and accuracy in results is tested using a Test dataset. This dataset has examples that the model hasn't seen during the training.The model makes prediction on this dataset and its performance is assessed by comparing these predictions to the true labels.  </font>**

**<font color = red > Support Vector Machine (SVM) algorithim is a supervised learning method used for classification and regression tasks.SVM creates a linear classification function for two class classification.In order to work, it first determines which feature space hyperplane best divides the classes.  Its objective is to maximize the margin which is the distance between the hyperplane and the closest data points from each class. By utilizing the kernel function, SVM can handle datasets that are both linearly and non-linearly separable, which is especially useful when working with high demensional data. </font>**

**<font color = red > A Kernel is a function that measures the similarity between the pairs of data points in a higher dimensional space. They enable the transformation of input data into a higher dimensional space in SVM without requiring the computation of the data points' new coordinates.  Because of this, SVM is able to identify complicated links and patterns in the data that might not have been visible in the original feature space. kernel has different types like linear, polynomial, or radial basis function (RBF), this range of  different types helps SVM to adapt to different data characteristics, enhancing its performance across diverse tasks. </font>**

**<font color = blue > 2. Perform k-means clustering, choosing two numerical features
in your dataset, and apply k-means clustering to your data to
create k clusters in Python (k>=2)</font>**

```python
year_1990=dataset2_wrangle['Life Expentancy 1990']
year_2010=dataset2_wrangle['Life Expentancy 2010']
```

```python
year_1990_df=pd.DataFrame(year_1990,columns=['Life Expentancy 1990'])
year_2010_df=pd.DataFrame(year_2010,columns=['Life Expentancy 2010'])
```

```python
Both_years=pd.concat([year_1990_df,year_2010_df],axis=1)
Both_years
```

```python
from sklearn.cluster import KMeans
```

```python
plt.scatter(x=dataset2_wrangle['Life Expentancy 1990'],y=dataset2_wrangle['Life Expentancy 2010'])
plt.xlabel('Life Expentancy 1990 (age)')
plt.ylabel('Life Expentancy 2010(age)')
plt.title('Life Expentancy in year 1990 against Life Expentancy in year 2010 across different countries'' multiple regions')
```

```python
kmeans = KMeans(n_clusters=2).fit(
 Both_years[['Life Expentancy 1990','Life Expentancy 2010']]
)
```

```python
plt.scatter(
 x=dataset2_wrangle['Life Expentancy 1990'],
 y=dataset2_wrangle['Life Expentancy 2010'],
 c=kmeans.labels_)
plt.plot(
 kmeans.cluster_centers_[:,0],
 kmeans.cluster_centers_[:,1],
 'k*',
 markersize=20,
 markerfacecolor='red'  
)
plt.xlabel('Life Expentancy in 1990 of people (age)')
plt.ylabel('Life Expentancy in 2010 of people (age)')
plt.title('Life Expentancy in year 1990 against Life Expentancy in year 2010 across different countries'' multiple regions')
plt.show()
```
